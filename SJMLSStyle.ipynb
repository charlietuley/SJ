```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
df = pd.read_csv('MLSStyle1 - Sheet1.csv')
```

```python
df
```

```python
features = ['SoT','npxG/Sh','PPA','CrsPA','%Live','%Dead','%Drib','Att 3rd','Att Pen']
x = df.loc[:, features].values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
pd.DataFrame(data = x, columns = features).head()
```

```python
pca = PCA(n_components=2)
```

```python
principalComponents = pca.fit_transform(x)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf
```

```python
principalDf['Euc Dist'] = np.sqrt(((principalDf['principal component 1']-1.443118)**2)+((principalDf['principal component 2']+0.212162)**2))
```

```python
principalDf
```

```python
teams = df['Squad']
principalDf = principalDf.join(teams)
```

```python
principalDf
```

```python
principalDf.sort_values(by=['Euc Dist'], inplace=True, ascending=False)
```

```python
principalDf
```

```python
#Most similar Attacking
#Houston,Orlando,Portland,Columbus,NYRB
```
